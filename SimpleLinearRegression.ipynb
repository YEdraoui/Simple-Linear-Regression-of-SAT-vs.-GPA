# Simple linear regression

## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('1.01.+Simple+linear+regression.csv')

# Define dependent and independent variables
y = data['GPA']
x1 = data['SAT']

# Reshape x1 for NumPy
x1 = x1.values.reshape(-1, 1)

# Scatter plot of SAT vs GPA (without seaborn)
plt.scatter(x1, y, color='blue')
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.title('Scatter plot of SAT vs GPA')
plt.show()

# Perform linear regression using NumPy (manual calculation)
X_b = np.c_[np.ones((x1.shape[0], 1)), x1]  # add bias (x0 = 1) to each instance
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)  # calculate theta

# Slope (theta1) and intercept (theta0)
slope = theta_best[1]
intercept = theta_best[0]

# Print the model's slope (coefficient) and intercept
print(f"Slope (coefficient): {slope}")
print(f"Intercept: {intercept}")

# Predict using the linear regression model
yhat = X_b.dot(theta_best)

# Plot the regression line
plt.scatter(x1, y, color='blue')
plt.plot(x1, yhat, lw=4, color='orange', label='Regression Line')
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.title('Regression Line for SAT vs GPA')
plt.show()
```
